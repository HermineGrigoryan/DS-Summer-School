## Merging

```python
df_dict={"make":["audi", "bmw"], "rating":[4,5]}
df1=pd.DataFrame(df_dict)
df1
```

```python
df2=pd.DataFrame(df_dict)
df2
```

```python
pd.concat([df1, df2], axis=0) #0 - rbind, 1 - cbind
```

* outer - union (showing all the columns)
* inner - intersection
* left - add the columns of the right one, leave the left one.
* right - add the columns of the left one, leave the right one.
* index
* key column

```python
data.merge(df1, how="inner")
data.merge(df1, how="outer")
data.merge(df1, how="left")
data.merge(df1, how="right")
```
